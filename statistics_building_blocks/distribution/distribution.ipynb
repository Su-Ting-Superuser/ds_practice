# PDF, PMF and CDF

- **PDF**, Probability Density Function, is the density of a **continuous random variable**. 
- **PMF**, Probability Mass Function, is a function that gives the **probability** that a **discrete random variable** is exactly **equal to some value**.
- **CDF**, is the **probability** that **X** will take a value **less than or equal to x**.

```python
import numpy as np
import seaborn as sns

data = np.random.randn(500)
```

```python
# PDF
sns.distplot(data)
```

```python
# CDF
kwargs = {'cumulative': True}
sns.distplot(data, hist_kws=kwargs, kde_kws=kwargs)
```
